### Question 1: What is a random variable in probability theory?

A random variable is a variable that represents the outcomes of a random experiment in numerical form. It assigns a real number to each possible outcome of a random event. For example, the number of heads in three coin tosses is a random variable.

### Question 2: What are the types of random variables?

There are two types of random variables:
1. Discrete Random Variable – takes countable values (e.g., number of students in a class).
2. Continuous Random Variable – takes infinite possible values within a range (e.g., height, weight)

### Question 3: Explain the difference between discrete and continuous distributions.

A discrete distribution describes outcomes with a finite or countable number of possibilities (e.g., binomial distribution). A continuous distribution deals with outcomes that can take any value within a range (e.g., normal distribution).

### Question 4: What is a binomial distribution, and how is it used in probability?

A binomial distribution represents the number of successes in a fixed number of independent Bernoulli trials, each with the same probability of success. It is used to model binary outcomes like yes/no or success/failure.

### Question 5: What is the standard normal distribution, and why is it important?

The standard normal distribution is a normal distribution with a mean of 0 and a standard deviation of 1. It is important because it simplifies statistical analysis, allowing the use of z-scores to compare data points from different normal distributions.

### Question 6: What is the Central Limit Theorem (CLT), and why is it critical in statistics?

The Central Limit Theorem states that the sampling distribution of the sample mean approaches a normal distribution as the sample size increases, regardless of the population's distribution. It allows us to make inferences and construct confidence intervals even when the population distribution is unknown.

### Question 7: What is the significance of confidence intervals in statistical analysis?

Confidence intervals provide a range of values within which the true population parameter is expected to lie with a certain probability (e.g., 95%). They indicate the reliability of an estimate.

### Question 8: What is the concept of expected value in a probability distribution?

The expected value of a random variable is the long-run average value it would take after many repetitions of the random experiment. It is calculated as the sum (or integral) of each possible value multiplied by its probability.

### Question 9: Write a Python program to generate 1000 random numbers from a normal distribution with mean = 50 and standard deviation = 5. Compute its mean and standard deviation using NumPy, and draw a histogram to visualize the distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(50, 5, 1000)
mean = np.mean(data)
std_dev = np.std(data)

print("Mean:", mean)
print("Standard Deviation:", std_dev)

plt.hist(data, bins=30, edgecolor='black')
plt.title('Normal Distribution (mean=50, sd=5)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

### Question 10: You are working as a data analyst for a retail company. The company has collected daily sales data for 2 years and wants you to identify the overall sales trend. daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,  235, 260, 245, 250, 225, 270, 265, 255, 250, 260] Explain how you would apply the Central Limit Theorem to estimate the average sales with a 95% confidence interval. Write the Python code to compute the mean sales and its confidence interval.

The Central Limit Theorem (CLT) states that if we take many random samples from a population, the sampling distribution of the sample mean will approximate a normal distribution, even if the original data are not perfectly normal — provided the sample size is reasonably large.

To estimate the average sales and build a 95% confidence interval, we:

Calculate the sample mean and standard deviation of daily sales.

Use the t-distribution (since the sample size is small, n = 20).

Compute the confidence interval (CI) using the formula. 



```python
import numpy as np
import scipy.stats as st

daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

mean_sales = np.mean(daily_sales)
std_err = st.sem(daily_sales)
conf_interval = st.t.interval(0.95, len(daily_sales)-1, loc=mean_sales, scale=std_err)

print("Mean Sales:", mean_sales)
print("95% Confidence Interval:", conf_interval)
```
